```python

T={}
order = 6

for ix in range(len(edited)-order):
    ctx = edited[ix:ix+order]
    futval = edited[ix+order]
    
    if T.get(ctx) is None:
        T[ctx] = {}
        T[ctx][futval] = 1
        
    else:
        if T[ctx].get(futval) is None: #ctx hai but futval ni? how?
            T[ctx][futval] = 1
        else:
            T[ctx][futval] +=1
        
    
for kx in T.keys():
    
    s = float(sum(T[kx].values()))
    
    for k in T[kx].keys():
        T[kx][k] = T[kx][k]/s
    
```

```python
def test_samp(probs, temp = 1.0):
    #array mei kro 
    probs = np.asarray(probs)
    exp_probs = np.exp(np.log(probs)/temp)
    
    return list(exp_probs/sum(exp_probs))

#list bhejooo 
```

```python
def gennextval(ctx,diversity = 1.0):
    if T.get(ctx) is None:
        return ' '
    else:
        divprobs = test_samp(T[ctx].values(), temp = diversity)
        return np.random.choice(T[ctx].keys(),p= divprobs)
    
```

```python
ind = int (np.random.random()*(len(edited)-order))
ctx = edited[ind:ind+order] 

sentance = ''+ctx

for im in range(2000):
    x = gennextval(ctx, diversity = 1.3)
    sentance += x 
    ctx = sentance[-order:]
    
print sentance
```
